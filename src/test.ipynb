```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
masks = np.load('masks.npy')
masks.shape
```

```python

# show bounding box (xmin, ymin, xmax, ymax)
bbox = [169, 167, 313, 423]

# show bounding box as a rectangle
plt.plot([bbox[1], bbox[1]], [bbox[0], bbox[2]], 'r-')
plt.plot([bbox[3], bbox[3]], [bbox[0], bbox[2]], 'r-')
plt.plot([bbox[1], bbox[3]], [bbox[0], bbox[0]], 'r-')
plt.plot([bbox[1], bbox[3]], [bbox[2], bbox[2]], 'r-')

xmin_px, ymin_px, xmax_px, ymax_px = bbox
top_mask = max(masks, key=lambda mask: mask[xmin_px:xmax_px, ymin_px:ymax_px].sum())

top_mask[top_mask] = 0
plt.imshow(top_mask)
top_mask.max()
```
